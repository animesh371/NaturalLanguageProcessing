```python
import nltk
from nltk.corpus import state_union
from nltk.tokenize import PunktSentenceTokenizer
```

```python
"""PunktSentenceTokenizer is a unsupervised machine learning tokenizer.
It comes pre-trained but can be retrained by feeding in the data again"""
```

```python
"""state_union is a collection of speeches given by the US President to a joint
session of the United States Congress"""
```

```python
train_text = state_union.raw("1999-Clinton.txt")
test_text = state_union.raw("2000-Clinton.txt")
```

```python
custom_sent_tokenizer = PunktSentenceTokenizer(train_text)
tokenized = custom_sent_tokenizer.tokenize(test_text)
```

```python
""" The below function takes a piece of text as input, tokenizes it and determines the 
part of speech of every word in the sentence"""
```

```python
def process_content():
    try:
        for i in tokenized:
            words = nltk.word_tokenize(i)
            tagged = nltk.pos_tag(words)
            print (tagged)
    except Exception as e:
        print (str(e))
```

```python
process_content()
```
